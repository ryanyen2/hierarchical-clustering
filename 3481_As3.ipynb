```python
from matplotlib import pyplot as plt

# give you a cluster result as predicted and also its counts
def getClusterInfo(nOfCluster,clusterInfo):
    kclusters = f(clusterInfo, nOfCluster, criterion='maxclust')

    # build a table to show the number of count
    clusterCount = []
    for i in range(0,nOfCluster):
        clusterCount.append(0)
    for i in range(0,len(kclusters)):
        clusterCount[kclusters[i]-1] += 1;
    for i in range(0,nOfCluster):
        print(str(i+1) + " : "+str(clusterCount[i])+'\n')
    return kclusters
```
